# Imbalanced Dataset

```python
#import libraraies
import numpy as np
import pandas as pd
```

```python
#Read the data from the file
df_train = pd.read_csv("/content/train.csv")
df_train
```

```python
#calculate the class distribution before resampling
target_count = df_train.target.value_counts()
print("class 0:",target_count[0])
print("class 1:",target_count[1])
print("portion of class 0:",round(target_count[0]*100/(target_count[1]+target_count[0]),2),'%')
```

**Splitting**

```python
#splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
labels = df_train.columns[2:]
df_train = df_train.dropna()

x = df_train[labels]
y = df_train['target']


```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
```

**Scaling the data**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

**Random undersampling**

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler()
x_rus,y_rus = rus.fit_resample(x_train_scaled,y_train)
print("before random sampling",y.value_counts)
print("after random sampling",pd.Series(y_rus).value_counts())


```

**Random Oversampling**

```python
#performing Random Over-Sampling to balance the minority class

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
x_ros,y_ros = ros.fit_resample(x_train_scaled,y_train)
print("before",y.value_counts())
print("after",pd.Series(y_ros).value_counts())
```

```python
#Displaying the total number of class 1 samples for RUS,ROS and ORIGINAL data

y_rus.sum(),y_ros.sum(),y.sum()
```

**UNDERSAMPLING : CLUSTER CENTROIDS**

```python
from imblearn.under_sampling import ClusterCentroids, RandomUnderSampler
import pandas as pd

cc = ClusterCentroids(sampling_strategy={0:100}, random_state=42)
X_cc, y_cc = cc.fit_resample(x_train_scaled, y_train)

print("\nBefore ClusterCentroids:", y.value_counts())
print("After ClusterCentroids:\n", pd.Series(y_cc).value_counts())
```

**OVERSAMPLING : SMOTE**

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

sm = SMOTE(random_state=42)
x_train_sm, y_train_sm = sm.fit_resample(x_train_scaled, y_train)

print("\nBefore SMOTE:", y.value_counts())
print("After SMOTE:\n", pd.Series(y_train_sm).value_counts())
```
